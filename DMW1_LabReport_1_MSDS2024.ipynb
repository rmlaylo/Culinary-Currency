<h4>Merge tables to unified table</h4>

Next is to merge each individual monthly Customs Imports table into one unified table. An additional column was first added to each monthly table to identify the month and year of where each import entry originated. Afterwards, all monthly tables were unified into the `customs_imports` table.

```python
# Add Month column to each monthly Customs Imports table
with sqlite3.connect(db_name) as conn:
    cursor = conn.cursor()
    for month in months:
        try:
            cursor.execute(
                f"""
                ALTER TABLE {month}
                ADD COLUMN Month;
                """)
            cursor.execute(
                f"""
                UPDATE {month}
                SET Month = ?""",
                (month,))
        except sqlite3.OperationalError as e:
            continue

# Join the twelve monthly tables into one unified Customs Imports Table
def unify_tables(db_name, unified_table_name, tables):
    """
    """
    with sqlite3.connect(db_name) as conn:
        cursor = conn.cursor()
        cursor.execute(
            f"""
            CREATE TABLE {unified_table_name} AS
            SELECT
                *
            FROM
                {tables[0]};
            """
        )
        for table in tables[1:]:
            cursor.execute(
                f"""
                INSERT INTO
                    {unified_table_name}
                SELECT
                    *
                FROM
                    {table};
                """
            )

unified_table_name = 'customs_imports'
unify_tables(db_name, unified_table_name, months)
```

<h4>Compute contribution of top ingredients per cuisine</h4>

In order to feasibly assign an `HSCODE` to every ingredient, the scope of the ingredients to prioritize and include in the study must be defined. Upon initial check of the dataset, a total number of more than 6,700 unique ingredients were identified. The next objective is to determine the appropriate number of ingredients that would be sufficiently representative per cuisine.

```python
# Check for count of unique ingredients
print('Number of unique ingredients: ', len(df_recipe['ingredients'].unique()))
```

Next, the ingredients were segregated per `cuisine`, and sorted based on the number of times it has appeared in a unique recipe for said `cuisine`. Top ingredients would be prioritized to be selected as part of the representative group per `cuisine`.

```python
# Segregrate ingredients per cuisine
# Sort based on count of unique recipes that use the ingredient
df_recipe_count = (pd.DataFrame(
    df_recipe.groupby(['cuisine', 'ingredients'])['ingredients'].count())
                   .reset_index(names=['cuisine', 'ingredient_key'])
                   .sort_values(['cuisine', 'ingredients'],
                                ascending=[True, False])
                   .rename(columns={'ingredients': 'recipe_count'})
                  )

df_recipe_count.head()
```

The cumulative percentage of the `recipe_count` of each ingredient versus the total `recipe_count` per `cuisine` was computed.

```python
# Compute the cumulative percentage of ingredients vs total count per cuisine
total_recipe_count = df_recipe.groupby(
    ['cuisine']).count().reset_index()[['cuisine', 'ingredients']]
df_temp = pd.merge(
    df_recipe_count, total_recipe_count, how='left', on='cuisine')
df_temp['perc_total'] = (df_temp['recipe_count'] / df_temp['ingredients'])
df_temp['cum_perc'] = df_temp.groupby('cuisine')['perc_total'].cumsum()
df_temp.columns = ['cuisine', 'ingredient_key', 'recipe_count',
                   'recipe_count_cuisine', 'perc2total', 'cum_perc']
df_temp.head()
```

Using the cumulative percentage of the `recipe_count` per `cuisine`, the average cumulative percentage per `cuisine` was then computed depending on the number of top ingredients included, ranging from top 10 to top 50. Based on results, each `cuisine` would need to include at least 50 ingredients. This would translate to a total of 312 unique ingredients that need to assigned a corresponding `HSCODE`.

```python
# Check top recipe count per cuisine needed to get ~50%
top_count_iter = [10, 15, 20, 25, 30, 35, 40, 45, 50]
for top_count in top_count_iter:
    df_temp_top = (df_temp
                   .groupby('cuisine')
                   .head(top_count)
                   .reset_index(drop=True))
    print(top_count, ':',
          df_temp_top.groupby('cuisine')['cum_perc'].max().mean())

# Compute the total number of unique ingredients for scope of top 50 per cuisine
df_top_recipe_count = df_temp.groupby('cuisine').head(50).reset_index(drop=True)
print('Number of unique ingredients:',
      len(df_top_recipe_count['ingredient_key'].unique()))
```

```python
# Segregrate ingredients per cuisine
# Sort based on count of unique recipes that use the ingredient
df_recipe_count = (pd.DataFrame(
    df_recipe.groupby(['cuisine', 'ingredients'])['ingredients'].count())
                   .reset_index(names=['cuisine', 'ingredient_key'])
                   .sort_values(['cuisine', 'ingredients'],
                                ascending=[True, False])
                   .rename(columns={'ingredients': 'recipe_count'})
                  )

# Compute the cumulative percentage of ingredients vs total count per cuisine
total_recipe_count = df_recipe.groupby(
    ['cuisine']).count().reset_index()[['cuisine', 'ingredients']]
df_temp = pd.merge(
    df_recipe_count, total_recipe_count, how='left', on='cuisine')
df_temp['perc_total'] = (df_temp['recipe_count'] / df_temp['ingredients'])
df_temp['cum_perc'] = df_temp.groupby('cuisine')['perc_total'].cumsum()
df_temp.columns = ['cuisine', 'ingredient_key', 'recipe_count',
                   'recipe_count_cuisine', 'perc2total', 'cum_perc']

# Check top recipe count per cuisine needed to get average of 50%
top_count_iter = [10, 15, 20, 25, 30, 35, 40, 45, 50]
for top_count in top_count_iter:
    df_temp_top = df_temp.groupby(
        'cuisine').head(top_count).reset_index(drop=True)
    print(top_count, ':',
          df_temp_top.groupby('cuisine')['cum_perc'].max().mean())

# Compute the total number of unique ingredients for scope of top 50 per cuisine
df_top_recipe_count = df_temp.groupby('cuisine').head(20).reset_index(drop=True)
print('Number of unique ingredients:',
      len(df_top_recipe_count['ingredient_key'].unique()))
```

<h4>Create Consolidated table</h4>

The final table to create is the `consolidated` table that will link the `customs_imports`, `recipe_ingredients` and `ingredients_hscode` together and will be utilized heavily for exploratory data analysis.

First, another table is created that will utilize the `ingredients_HSCODE` table and determine all row entries found in `customs_imports` that satisfy these two conditions:
1. `HSCODE` in both tables match with each other
2. The `regex_keyword` in `ingredients_HSCODE` is contained in the `GOODSDESCRIPTION` column from the `customs_imports` table

After which, each row entry that satisfies those two conditions are joined to the corresponding `ingredients_HSCODE` entry.

```python
# Create a table that identifies all valid customs_imports entries
with sqlite3.connect(db_name) as conn:
    cursor = conn.cursor()
    cursor.execute(
        """
        DROP TABLE IF EXISTS ingred_customs
        """
    )
    cursor.execute(
        """
        CREATE TABLE ingred_customs AS
        SELECT
            *
        FROM
            ingredients_hscode i
        LEFT JOIN
            customs_imports m
        ON
            i.HSCODE = m.HSCODE
        WHERE
            GOODSDESCRIPTION LIKE regex_keyword
        """
    )
```

Finally, the `consolidated` table is created by joining the `ingred_customs` table and the `recipe_ingredients` table. The target is to link to each cuisine-ingredient pair all corresponding customs row entries that match according to the identified `ingredient_key`.

```python
with sqlite3.connect(db_name) as conn:
    cursor = conn.cursor()
    cursor.execute(
        """
        DROP TABLE IF EXISTS consolidated
        """
    )
    cursor.execute(
        """
        CREATE TABLE consolidated AS
        SELECT
            Month, r.cuisine, r.ingredient_key, GOODSDESCRIPTION, i.HSCODE,
            COUNTRYEXPORT, GROSSMASSKGS, NETMASSKGS, CUSTOMSVALUE,
            CURRENCY, DUTIABLE_FOREIGN, EXCHANGE_RATE, DUTIABLEVALUEPHP,
            FREIGHT, INSURANCE, ARRASTRE, VATTAXBASE, DUTIESTAXES,
            recipe_count, recipe_count_cuisine
        FROM
            recipe_ingredients r
        LEFT JOIN
            ingred_customs i
        ON
            r.ingredient_key = i.ingredient_key
        """
    )
```

<h4>Compute for the total costs per cuisine and their percent contribution</h4>

Firstly the total import costs were defined as the following relevant variables:
- `INSURANCE` (in native currency)
- `ARRASTRE` (in native currency)
- `VATTAXBASE` (in Philippine Peso)
- `DUTIESTAXES` (in Philippine Peso)

The `INSURANCE` and `ARRASTRE` would then be multiplied by the `EXCHANGE_RATE` variable to convert it to Philippine Peso.

```python
# Compute for total costs per cuisine, percent distribution
df_cost_total = pd.read_sql(
    """
    WITH TotalCosts AS (
        SELECT
            cuisine,
            SUM(INSURANCE * EXCHANGE_RATE + ARRASTRE * EXCHANGE_RATE
            + VATTAXBASE + DUTIESTAXES) AS TOTAL_COSTS
        FROM consolidated
        GROUP BY cuisine
    )
    SELECT
        TC.cuisine,
        TC.TOTAL_COSTS,
        (TC.TOTAL_COSTS / SUM(TC.TOTAL_COSTS) OVER ()) * 100
            AS TOTAL_COST_PERCENT
    FROM TotalCosts AS TC
    ORDER BY
        TOTAL_COST_PERCENT desc;
    """,
    conn)

df_cost_total.head(10)
```

Based on the results of the analysis, `Filipino` cuisine ranks the highest amongst other cuisines based on total import cost, at <b>9.19%</b> percent distribution, with a significant difference from the next highest cuisine (`Southern US` at 6.76%).

<h4>Compute for total costs per kgs cuisine and their percent distribution</h4>

Next, using the same calculation to compute for the total import costs earlier, the costs per kgs is calculated by dividing the summation with two variables: `GROSSMASSKGS` and `NETMASSKGS`, and taking the percentage distribution.

```python
# Compute for total costs per kgs cuisine, percent distribution
df_mass_total = pd.read_sql(
    """
    WITH TotalCosts AS (
        SELECT
            cuisine,
            SUM(INSURANCE * EXCHANGE_RATE + ARRASTRE * EXCHANGE_RATE
            + VATTAXBASE + DUTIESTAXES) AS TOTAL_IMPORT_COST,
            SUM(GROSSMASSKGS) AS TOTAL_GROSS_MASS,
            SUM(NETMASSKGS) AS TOTAL_NET_MASS
        FROM
            consolidated
        GROUP BY
            cuisine
    )
    SELECT
        TC.cuisine,
        TC.TOTAL_IMPORT_COST / TC.TOTAL_GROSS_MASS AS COST_PER_GROSSMASS,
        (TC.TOTAL_GROSS_MASS / SUM(TC.TOTAL_GROSS_MASS) OVER ()) * 100
        AS COST_PER_GROSSMASS_PERCENT,
        TC.TOTAL_IMPORT_COST / TC.TOTAL_NET_MASS AS COST_PER_NETMASS,
        (TC.TOTAL_GROSS_MASS / SUM(TC.TOTAL_GROSS_MASS) OVER ()) * 100
        AS COST_PER_NETMASS_PERCENT
    FROM
        TotalCosts AS TC
    ORDER BY
        COST_PER_NETMASS_PERCENT desc;
    """, conn)

df_mass_total.head(10)
```

Once again, the `Filipino` cuisine ranks the highest amongst other cuisines based on the total import costs per kgs, both for `GROSSMASSKGS` and `NETMASSKGS`.

<h4>Identify top 5 ingredients for top cuisine</h4>

Given that the top costly cuisine is `Filipino`, the next step is to identify which of the cuisine's ingredients constitute as the most expensive to import based on the cost per kgs. Specifically we will be utilizing primarily the `NETMASSKGS` as the denominator in order to take into consideration the difference that includes packaging. The rational behind this is due to transportation costs involving the goods would likely taking into consideration the costs inclusive of the packaging it was imported with.

```python
# Identify top 5 ingredients for top cuisine (Filipino)
df_top5_ing_top1_cuisine = pd.read_sql(
    """
    WITH RankedIngredients AS (
        SELECT
            cuisine,
            ingredient_key,
            SUM((INSURANCE * EXCHANGE_RATE + ARRASTRE * EXCHANGE_RATE
            + VATTAXBASE + DUTIESTAXES) / NETMASSKGS) AS COST_PER_NETMASS,
        RANK() OVER (
            PARTITION BY
                cuisine
            ORDER BY
                SUM((INSURANCE * EXCHANGE_RATE + ARRASTRE * EXCHANGE_RATE +
                VATTAXBASE + DUTIESTAXES) / NETMASSKGS) DESC)
                    AS ingredient_rank
        FROM consolidated
        GROUP BY
            cuisine,
            ingredient_key
        )
    SELECT
        cuisine,
        ingredient_key,
        COST_PER_NETMASS
    FROM
        RankedIngredients
    WHERE
        ingredient_rank <= 5 AND 
        cuisine LIKE 'filipino'
    ORDER BY
        COST_PER_NETMASS desc;
    """,
    conn)

df_top5_ing_top1_cuisine
```

<h4>Identify the most costly ingredients for all cuisines combined</h4>

The top costly ingredients among all cuisines and compared to the ranking of pork frequently used in Filipino cuisine.

```python
# Check for top ingredients in general (regardless of cuisine)
df_top5_ing_per_cuisine = pd.read_sql(
    """
    WITH RankedIngredients AS (
        SELECT
            cuisine,
            ingredient_key,
            SUM((INSURANCE * EXCHANGE_RATE + ARRASTRE * EXCHANGE_RATE
            + VATTAXBASE + DUTIESTAXES) / NETMASSKGS) AS COST_PER_NETMASS,
        RANK() OVER (
            PARTITION BY
                cuisine
            ORDER BY
                SUM((INSURANCE * EXCHANGE_RATE + ARRASTRE * EXCHANGE_RATE +
                VATTAXBASE + DUTIESTAXES) / NETMASSKGS) DESC)
                    AS ingredient_rank
        FROM consolidated
        GROUP BY
            cuisine,
            ingredient_key
        )
    SELECT
        cuisine,
        ingredient_key,
        COST_PER_NETMASS
    FROM
        RankedIngredients
    WHERE
        ingredient_rank <= 5
    ORDER BY
        COST_PER_NETMASS desc;
    """,
    conn)

df_top5_ing_per_cuisine.head(15)
```

Pork is ranked at the top and with a cost per kgs cost significantly higher than the remaining top 15 ingredients displayed, validating the presumption that pork is a generally expensive ingredient to import. As a side note, beef popularly used for Korean cuisine ranked second among ingredients, with Korean ranking top 3 among other cuisines for overall cost, and top 8 for cost per kgs, highlighting the significant factor that ingredient prices play to the cuisine's perceived cost.

<h4>Top Exporter Countries for Pork (Volume)</h4>

```python
# Identify top exporter countries for pork (volume)
df_exporters_volume = pd.read_sql(
    """
    WITH IngredientVolume AS (
        SELECT
            cuisine,
            ingredient_key,
            COUNTRYEXPORT,
            SUM(NETMASSKGS) AS TOTAL_MASS
        FROM
            consolidated
        WHERE
            ingredient_key = 'pork'
        GROUP BY
            cuisine, ingredient_key, COUNTRYEXPORT
    )
    SELECT
        cuisine,
        ingredient_key,
        COUNTRYEXPORT,
        TOTAL_MASS,
        (TOTAL_MASS / SUM(TOTAL_MASS) OVER ()) * 100
            AS TOTAL_MASS_PERCENT
    FROM
        IngredientVolume
    ORDER BY
        TOTAL_MASS_PERCENT desc;
    """,
    conn)

df_exporters_volume.head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.barh(df_exporters_volume['COUNTRYEXPORT'],
         df_exporters_volume['TOTAL_MASS_PERCENT'], color='tomato')
plt.xlabel('Total Net Mass (%)')
plt.ylabel('Country Export')
plt.title('Figure 4. Ranked Pork Exporters by Total Mass Percentage')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

Based on the analysis, Spain was identified as the top exporter of pork to the Philippines in terms of volume, contributing <b>24%</b> of overall pork importation to the Philippines. Canada and Brazil rank 2nd and 3rd with a percentage of 18% and 16% respectively.

<h4>Top Exporter Countries for Pork (Cost per kgs)</h4>

```python
# Identify top exporter countries for pork (cost per kgs)
df_exporters_avg_cost_per_kg = pd.read_sql(
    """
    WITH IngredientCosts AS (
        SELECT
            cuisine,
            ingredient_key,
            COUNTRYEXPORT,
            AVG((INSURANCE * EXCHANGE_RATE + ARRASTRE * EXCHANGE_RATE +
            VATTAXBASE + DUTIESTAXES) / NETMASSKGS)
                AS AVG_COST_PER_KG
        FROM
            consolidated
        WHERE
            ingredient_key = 'pork'
        GROUP BY
            cuisine, ingredient_key, COUNTRYEXPORT
    )
    SELECT
        cuisine,
        ingredient_key,
        COUNTRYEXPORT,
        AVG_COST_PER_KG
    FROM
        IngredientCosts
    ORDER BY
        AVG_COST_PER_KG desc;
    """,
    conn)

df_exporters_avg_cost_per_kg.head(20)
```

```python
plt.figure(figsize=(12, 6))
plt.barh(df_exporters_avg_cost_per_kg['COUNTRYEXPORT'],
         df_exporters_avg_cost_per_kg['AVG_COST_PER_KG'], color='tomato')
plt.xlabel('Average Cost per Kilogram')
plt.ylabel('Country Export')
plt.title('Figure 5. Ranked Pork Exporters by Average Cost per Kilogram')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
